## 1.2. ¿Qué es PyCaret?

**PyCaret** es una biblioteca de aprendizaje automático de bajo código en Python, diseñada para simplificar y acelerar el ciclo de vida de los modelos de machine learning. PyCaret permite a los usuarios construir, entrenar, ajustar y desplegar modelos con pocas líneas de código, ofreciendo una experiencia de prototipado rápido ideal para experimentación y desarrollo.

PyCaret es especialmente útil para usuarios que buscan:
- Realizar experimentación rápida en diferentes tipos de problemas (**clasificación, regresión, clustering, detección de anomalías, etc.**).
- Reducir el tiempo necesario para **seleccionar, evaluar y optimizar modelos**.
- Acceder a una interfaz intuitiva para pruebas y validación sin necesidad de escribir mucho código.

## 2.1. Flujo de Trabajo en Aprendizaje Supervisado

En aprendizaje supervisado, tenemos una variable objetivo conocida. El flujo de trabajo típico en PyCaret es:
Aquí tienes el flujo de trabajo en aprendizaje supervisado de PyCaret con emojis para hacerlo más visual y atractivo:

### 🌟 Flujo de Trabajo Típico en Aprendizaje Supervisado con PyCaret

1. **⚙️ Setup**: Configura el entorno de trabajo, define el tipo de problema y la variable objetivo. Aquí se hace el preprocesamiento de los datos: codificación de variables categóricas, manejo de valores nulos y división en entrenamiento y prueba.

   ```python
   from pycaret.classification import *  # o pycaret.regression para regresión
   clf = setup(data=df, target='variable_objetivo', session_id=123)
   ```

2. **📈 Create Model**: Crea y entrena un modelo específico o varios para comparación. PyCaret tiene una variedad de algoritmos supervisados listos para usar (por ejemplo, regresión logística, árboles de decisión, bosques aleatorios).

   ```python
   best_model = create_model('rf')  # 'rf' es el Random Forest como ejemplo
   ```

3. **🔧 Tune Model**: Ajusta los hiperparámetros del modelo para mejorar su precisión y rendimiento. Este paso es crucial para optimizar el modelo y sacarle el máximo provecho.

   ```python
   tuned_model = tune_model(best_model)
   ```

4. **📊 Plot Model**: Genera gráficos para analizar el rendimiento del modelo. Puedes ver la matriz de confusión, la curva ROC y más, para evaluar la eficacia de las predicciones del modelo.

   ```python
   plot_model(tuned_model, plot='confusion_matrix')
   ```

5. **🧪 Evaluate Model**: PyCaret ofrece una interfaz de evaluación completa donde puedes analizar distintas métricas de rendimiento. Esto permite profundizar en el análisis y mejorar la comprensión del modelo.

   ```python
   evaluate_model(tuned_model)
   ```

6. **🔮 Predict Model**: Realiza predicciones en nuevos datos usando el modelo entrenado y optimizado.

   ```python
   predictions = predict_model(tuned_model, data=nuevos_datos)
   ```

7. **💾 Save Model**: Guarda el modelo para su uso posterior. Esto es útil para guardar el trabajo realizado y poder reutilizar el modelo sin necesidad de reentrenarlo.

   ```python
   save_model(tuned_model, 'mejor_modelo')
   ```

## Paso 1: Cargar Librerías y el Dataset 📥

Primero, cargamos las librerías necesarias y el conjunto de datos `insurance.csv`.

```python
import pandas as pd
from pycaret.regression import *

data.head()
```

## Paso 2: Configurar el Entorno de PyCaret 🛠️

Usamos la función `setup()` de PyCaret para configurar el entorno y definir la variable objetivo (`charges`).

```python
# Configurar PyCaret para el modelo de regresión
setup(data=data, target='charges', session_id=123)
```

> **🔵 Nota**: `session_id` se establece para garantizar resultados reproducibles.

## Paso 3: Comparar Modelos 🏆

Comparar todos los modelos de regresión disponibles y seleccionar el mejor en función de las métricas de rendimiento.

```python
# Comparar los modelos y encontrar el mejor
best_model = compare_models()
```

> **🟢 Consejo**: `compare_models()` genera una tabla con diferentes modelos y sus métricas de rendimiento, como RMSE y R2, lo que permite identificar el mejor modelo de forma rápida.

## Paso 4: Crear el Modelo Seleccionado 🧩

Creamos el modelo seleccionado para usarlo en el entrenamiento y prueba.

```python
# Crear el mejor modelo

Ajustamos los hiperparámetros del modelo seleccionado para mejorar su rendimiento.

```python
# Optimizar el modelo seleccionado
tuned_model = tune_model(model)
```

> **🟡 Importante**: `tune_model()` ajusta automáticamente los parámetros del modelo y selecciona los mejores valores para mejorar el rendimiento.

## Paso 6: Evaluar el Modelo 📊

Evaluamos el modelo con gráficos y métricas para verificar su precisión y otros aspectos de rendimiento.

```python
# Evaluar el modelo

> **🔴 Alerta**: Revisar los gráficos de errores y residuales ayuda a comprender mejor la precisión y ajustes del modelo.

## Paso 7: Realizar Predicciones 🧮

Utilizamos el modelo ajustado para realizar predicciones sobre un conjunto de datos nuevo o sobre el conjunto de entrenamiento.

```python

## 2.2. Flujo de Trabajo en Aprendizaje No Supervisado

En aprendizaje no supervisado, no tenemos una variable objetivo definida y buscamos patrones dentro de los datos, como clusters o anomalías.
Un flujo de trabajo típico en el módulo no supervisado de PyCaret consiste en seguir estos 6 pasos en este orden:

**Setup** ➡️ **Create Model** ➡️ **Assign Labels** ➡️ **Analyze Model** ➡️ **Prediction** ➡️ **Save Model**

1. **Setup**: Inicialización del entorno para clustering o detección de anomalías.
   ```python
   from pycaret.clustering import *
   clf = setup(data=df)
   ```

2. **Create Model**: Creación y entrenamiento de un modelo de clustering o detección de anomalías.
   ```python
   kmeans = create_model('kmeans')  # Por ejemplo, un modelo de K-Means
   ```

   ```python
   labeled_data = assign_model(kmeans)
   ```

   ```python
   plot_model(kmeans, plot='cluster')
   ```

5. **Evaluate Model**: Evaluación y análisis de los resultados del modelo.
   ```python
   evaluate_model(kmeans)
   ```

6. **Save Model**: Guardado del modelo de clustering para futuros análisis.
   ```python
   save_model(kmeans, 'kmeans_clustering_model')

# 🧑‍💻 Ejemplo Práctico en PyCaret: Detección de Anomalías en 7 Pasos

Este ejemplo práctico utiliza PyCaret para realizar un análisis no supervisado de detección de anomalías, identificando datos atípicos en un conjunto de datos. Este tipo de análisis es útil en aplicaciones como la detección de fraudes, mantenimiento predictivo y monitoreo de calidad.

## Paso 1: Cargar Librerías y el Dataset 📥

Primero, cargamos las librerías necesarias y el conjunto de datos para el análisis.

```python
import pandas as pd
from pycaret.anomaly import *

# Cargar el dataset de ejemplo
data = pd.read_csv("insurance.csv")
data.head()
```

## Paso 2: Configurar el Entorno de PyCaret para Anomalías 🛠️

Usamos la función `setup()` de PyCaret para configurar el entorno. Dado que es un análisis no supervisado, no necesitamos especificar una variable objetivo.

```python
# Configurar PyCaret para la detección de anomalías
setup(data=data, session_id=123)
```

> **🔵 Nota**: `session_id` garantiza la reproducibilidad, y `setup()` prepara los datos para el análisis.

## Paso 3: Crear el Modelo de Detección de Anomalías 🏆

Creamos un modelo de detección de anomalías. En este ejemplo, usamos el modelo `IForest` (Isolation Forest), que es popular en la detección de anomalías.

```python
# Crear un modelo de detección de anomalías (Isolation Forest)
model = create_model('iforest')
```

> **🟢 Consejo**: `IForest` es efectivo para detectar datos atípicos en grandes conjuntos de datos, basado en el aislamiento de puntos inusuales.

## Paso 4: Ajustar el Modelo 🔧

Si se desea, podemos optimizar algunos parámetros del modelo para mejorar su precisión en la detección de anomalías.

```python
# Optimización de parámetros (si es necesario)
tuned_model = tune_model(model)
```

> **🟡 Importante**: La optimización mejora el rendimiento del modelo, pero no siempre es necesaria en detección de anomalías.

## Paso 5: Etiquetar las Anomalías 📊

Usamos el modelo ajustado para etiquetar las observaciones en el conjunto de datos. Las observaciones se etiquetarán como "1" si se consideran anómalas, o "0" si son normales.

```python
# Etiquetar anomalías en el conjunto de datos original
anomaly_labels = predict_model(tuned_model, data=data)
anomaly_labels.head()
```

> **🔴 Alerta**: Estas etiquetas pueden ser usadas para analizar o filtrar solo los datos que se consideran anómalos.

## Paso 6: Evaluar el Modelo de Anomalías 📈

Evaluamos el modelo utilizando gráficos que muestren la distribución y localización de las anomalías.

```python
# Gráfico de anomalías
plot_model(tuned_model, plot='tsne')
```

> **🟢 Observación**: El gráfico `tsne` muestra la separación de los puntos anómalos y normales, permitiendo visualizar la efectividad del modelo en la detección de anomalías.

## Paso 7: Extraer y Analizar las Anomalías 🧮

Extraemos los datos etiquetados como anomalías para su análisis posterior.

```python
# Filtrar y mostrar solo las anomalías
anomalies = anomaly_labels[anomaly_labels['Anomaly'] == 1]
anomalies.head()
```

> **🔵 Resultado**: Este paso muestra los datos considerados anómalos, permitiendo un análisis detallado de estos casos.

Estos 7 pasos te permiten identificar y analizar anomalías en un conjunto de datos usando PyCaret. Esta metodología es útil en diversas aplicaciones donde es crucial detectar datos fuera de lo común.

````python


# Importar PyCaret para clustering, pandas para manipulación de datos
from pycaret.clustering import *
import pandas as pd

# Cargar el conjunto de datos de enfermedades cardíacas
from pycaret.datasets import get_data
data = get_data('heart_disease')

# Configurar PyCaret para clustering
cluster_setup = setup(data, session_id=123, normalize=True, ignore_features=['Disease'])

# Crear el modelo de clustering KMeans
kmeans_model = create_model('kmeans', num_clusters=3)  # Ajusta el número de clusters según sea necesario

# Función para solicitar al usuario sus datos de salud y predecir su cluster
def predict_cluster(model):
    print("Por favor, ingresa los siguientes datos de salud:")

    # Solicitar datos del usuario
    age = int(input("Edad: "))
    sex = int(input("Sexo (1=Masculino, 0=Femenino): "))
    cp = int(input("Tipo de dolor en el pecho (0-3): "))
    trestbps = int(input("Presión arterial en reposo (mm Hg): "))
    chol = int(input("Colesterol sérico (mg/dl): "))
    fbs = int(input("Azúcar en sangre en ayunas (1=Si, 0=No): "))
    restecg = int(input("Resultados del electrocardiograma en reposo (0-2): "))
    thalach = int(input("Frecuencia cardíaca máxima alcanzada: "))
    exang = int(input("¿Angina inducida por el ejercicio? (1=Sí, 0=No): "))
    oldpeak = float(input("Depresión del ST: "))
    slope = int(input("Pendiente del ST (0-2): "))
    ca = int(input("Número de vasos coloreados por fluoroscopia (0-3): "))
    thal = int(input("Tipo de talasemia (1-3): "))

    # Crear un DataFrame con los nombres de columna correctos
    user_data = pd.DataFrame({
        'age': [age],
        'sex': [sex],
        'chest pain type': [cp],
        'resting blood pressure': [trestbps],
        'serum cholestoral in mg/dl      ': [chol],
        'fasting blood sugar > 120 mg/dl ': [fbs],
        'resting electrocardiographic results': [restecg],
        'maximum heart rate achieved  ': [thalach],
        'exercise induced angina    ': [exang],
        'oldpeak ': [oldpeak],
        'slope of peak': [slope],
        'number of major vessels ': [ca],
        'thal': [thal]
    })

    # Asignar al usuario a un cluster
    prediction = predict_model(model, data=user_data)
    user_cluster = prediction['Cluster'][0]
    print(f"\nEl usuario pertenece al cluster: {user_cluster}")

    return user_cluster

# Realizar la predicción de cluster
usuario_cluster = predict_cluster(kmeans_model)

# Análisis de clusters para determinar niveles de salud
clustered_data = assign_model(kmeans_model)
cluster_health = clustered_data.groupby('Cluster').mean()

# Mostrar resultados de salud por cluster para comparar perfiles de riesgo
print("\nPromedio de características de salud por cluster (valores más bajos en colesterol, presión, etc. pueden indicar mejor salud):")
print(cluster_health)

# Evaluación de salud por cluster
print("\nInterpretación general:")
for cluster, stats in cluster_health.iterrows():
    if stats['serum cholestoral in mg/dl      '] < clustered_data['serum cholestoral in mg/dl      '].mean() and stats['resting blood pressure'] < clustered_data['resting blood pressure'].mean():
        print(f"Cluster {cluster} tiende a representar perfiles más saludables (colesterol y presión arterial más bajos).")
    else:
        print(f"Cluster {cluster} tiende a representar perfiles menos saludables (valores más altos en colesterol o presión arterial).")
````
